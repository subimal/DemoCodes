```python
import numpy as np
import scipy as sp
from scipy.fft import fft, ifft, fftfreq, fftshift, ifftshift
import pylab as pl

%matplotlib inline
```

```python
L = 30
xmin, xmax = -L/2, L/2
Nx = 2**13
dx = (xmax-xmin)/Nx
x = np.linspace(xmin, xmax, num=Nx, endpoint=False)
y = np.exp(1j*(x-5)) + np.exp(1j*(x-10)*4)

y_fft = fftshift(fft(y))
freq = np.pi/x #fftshift(fftfreq(x.shape[-1]))
```

```python
pl.figure()
pl.subplot(121)
pl.plot(x, y.real)
pl.subplot(122)
pl.plot(freq, np.abs(y_fft), '.-')
```

# $F[f(x)g(x)]$

```python
def f(x, w = 1):
    return (np.abs(x)<w/2)*1

def g(x, a = 1, b = 1):
    return 0*(x<0) + (1-x/a)*(x>0)*(x<=a)+ 0*(x>a)
        
```

```python
x = np.arange(-3,3, .01)
pl.figure()
pl.subplot(121)
pl.plot(x,f(x, w=1))
pl.plot(x,g(x))
pl.subplot(122)
pl.plot(x, np.abs(ifftshift(ifft(  fftshift(fft(f(x)))*fftshift(fft(g(x)))   ))) )
```
